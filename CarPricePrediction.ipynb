```python
import warnings
warnings.filterwarnings('ignore')

#Making new cloumn from existing column
data['year_diff'] = 2020 - data['Model_year']
```

```python
#Removing the model_year column as it is already transformed
data.drop("Model_year", inplace=True, axis=1)
```

```python
data['Price'] = data['Price'].str[4:]
```

```python
data['Price'] = data['Price'].astype("float32")

#Normalizing the price column
data['Price'] = data['Price']/100000
```

## Pycaret for Regression

```python
from pycaret.regression import *
```

```python
#Setting up environment
reg_setup = setup(data, target='Price', normalize=True, remove_outliers=True)
```

```python
#Comparing different models
compare_models()
```

```python
#As seen above Gradient Boosting Regressor has performed the best and linear regression has performed the worst
```

```python
'''
Using Gradient Boosting regression for prediction
'''
gbr = create_model('gbr',fold=10)
```

```python
evaluate_model(gbr)
```
